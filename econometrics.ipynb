```python
x = smf.add_constant(x)  # and add to regressors constant == betta_0*(A + u)
est = smf.OLS(y, x, missing='raise').fit()
est.summary()
```

```python
est = smf.OLS(y, x, missing='raise').fit(cov_type="HC3")  # боремся с гетероскедастичностью
est.summary()
# HC0 - корректировка стандартных ошибок Уайта
# Stata/Eviews - default is HC1
# (неудачно в питоне, но для малого набора данных сработает) HC3
```

```python
hypos = 'Занятые в тыс чел=0'  # checking the hypothesis that our constant = 0
est.t_test(hypos)
```

#** Heteroskedasticity tests**

##**Breush-Pagan test:**
Нулевая гипотеза в том, что var(u_i|X_i) = const отвергается, если F_act > F_{1-alpha}
Также можно проверить через p-value:
Если P(F)< alpha, где alpha  - уровень значимости, то гетероскедастичность значима, в противном случае - нет. У нас мы взяли alpha = 5%

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip

name = ["F actual", "p-value"]
test = sms.het_goldfeldquandt(est.resid, est.model.exog)
lzip(name, test)  
```

We have that p-value > alpha => Ошибка гомоскедастична

```python
from  statsmodels.stats.stattools import durbin_watson
durbin_watson(resids=est.resid)
```

Это значение близко к 2, значит автокорреляции нет
